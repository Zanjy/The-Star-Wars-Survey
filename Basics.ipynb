```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars = star_wars.loc[star_wars["RespondentID"].notnull(),:]
star_wars.head(10)
```

```python
star_wars.shape
```

## Cleaning and Mapping Yes/No Columns

```python
(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"]
 .value_counts(dropna = False))
```

```python
(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]
 .value_counts(dropna = False))
```

```python
q1 = "Have you seen any of the 6 films in the Star Wars franchise?"
q2 = "Do you consider yourself to be a fan of the Star Wars film franchise?"
```

```python
# Convert q1 & q2 columns to the Boolean type.
yes_no = {
    "Yes": True,
    "No": False
}

star_wars[q1] = star_wars[q1].map(yes_no)
star_wars[q2] = star_wars[q2].map(yes_no)
```

```python
star_wars[q1].head()
```

```python
star_wars[q2].head()
```

## Cleaning and Mapping Checkbox Columns

```python
star_wars.columns[3:9]
```

```python
# Convert each column on columns[3:9] so that it only contains the values True and False.
import numpy as np

yes_no_episode = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(yes_no_episode)
```

```python
# Rename each of the columns above so the names are more intuitive

star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "Seen_2",
    "Unnamed: 5": "Seen_3",
    "Unnamed: 6": "Seen_4",
    "Unnamed: 7": "Seen_5",
    "Unnamed: 8": "Seen_6",
})
star_wars.head()
```

## Cleaning the Ranking Columns

```python
# Convert each of the column in columns[9:15] to a float type.
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6"
})
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]].dtypes
```

## Finding the Most Viewed Movie

### Most Seen

```python
#Let's compute the sum of each of the seen columns & plot a bar chart
import matplotlib.pyplot as plt
%matplotlib inline

star_wars[star_wars.columns[3:9]].sum()
```

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
# plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

### The ranking

```python
# Let's compute the sum of each of the rank columns & plot a bar chart
star_wars[star_wars.columns[9:15]].mean()
```

```python
# bar chart of each ranking
star_wars[star_wars.columns[9:15]].mean().plot.bar()
```

We computed the average ranking of each movie, it looks like the "original" movies are rated much more highly than the newer ones. But the new ones are the most seen.


* "Star Wars: Episode V The Empire Strikes Back" is the most seen movie & least popular in term of ranking
* "Star Wars: Episode III  Revenge of the Sith" has the best ranking but also is the least seen 
* The most recent movies are the most seen
* The most seen correspond to the least popular in term of ranking
* And the least seen correspond to the most popular

## Exploring the Data by Binary Segments

```python
# We split the data into two groups : Male & Female
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

### The most seen by sex

```python
ax1 = males[males.columns[3:9]].sum().plot(kind="bar")
ax1.set_title("Male")
plt.show()

ax2 = females[females.columns[3:9]].sum().plot(kind="bar")
ax2.set_title("Female")
plt.show()
```

### The highest-ranked movie by sex

```python
ax1 = males[males.columns[9:15]].sum().plot(kind="bar")
ax1.set_title("Male")
plt.show()

ax2 = females[females.columns[9:15]].sum().plot(kind="bar")
ax2.set_title("Female")
plt.show()
```

> Interestingly, more males watches episodes 4, but males liked them far less than females did.
